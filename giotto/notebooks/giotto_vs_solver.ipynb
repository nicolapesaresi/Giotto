```python
from IPython.display import clear_output

from giotto.agents.alphazero import AlphaZeroAgent
from giotto.agents.bitbully import BitBullyAgent
from giotto.agents.minimax import MinimaxAgent
from giotto.envs.connect4 import Connect4Env
from giotto.envs.tris import TrisEnv
```

# Simulate a Giotto vs Solver game

### TicTacToe - solver: Minimax agent

```python
env = TrisEnv()
env.reset()  # random starter

agents = [MinimaxAgent(), AlphaZeroAgent(name="Giotto", game="tris", simulations=100, cpuct=1.5)]

starter_name = agents[env.current_player].name
second_name = agents[1 - env.current_player].name

env.render()
while not env.done:
    action = agents[env.current_player].select_action(env)
    env.step(action)
    clear_output()  # comment to print every move
    print(f"X player (first): {starter_name}")
    print(f"O player (second): {second_name}")
    env.render()

if env.info["winner"] == -1:
    print("Game over - Draw.")
else:
    print(f"Game over - {agents[env.info['winner']].name} wins.")
```

### Connect4 - solver: BitBully
