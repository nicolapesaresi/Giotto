```python
from IPython.display import clear_output

from giotto.agents.alphazero import AlphaZeroAgent
from giotto.agents.human import HumanAgent
from giotto.envs.connect4 import Connect4Env
from giotto.envs.tris import TrisEnv
```

# Play a game against Giotto

### TicTacToe

```python
env = TrisEnv()
env.reset()  # random starter

# [X, O]
agents = [HumanAgent(), AlphaZeroAgent(name="Giotto", game="tris", simulations=100, cpuct=1.5)]

env.render()
while not env.done:
    action = agents[env.current_player].select_action(env)
    env.step(action)
    clear_output()
    env.render()

if env.info["winner"] == -1:
    print("Game over - Draw.")
else:
    print(f"Game over - {agents[env.info['winner']].name} wins.")
```
