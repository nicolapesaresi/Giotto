```python
import json

from tqdm import tqdm

from giotto.agents.minimax import MinimaxAgent
from giotto.utils.random_board import random_board_connect4, random_board_tris
```

TicTacToe

```python
solver = MinimaxAgent()
```

```python
state, env = random_board_tris()
env.render()
print(f"To move: {env.signs[env.current_player]}")

print(solver.evaluate_all_moves(env))
print(solver.select_action(env))
```

```python
n_positions = 1000
filename = "tris_positions.jsonl"

with open(filename, mode="a") as f:
    for _ in tqdm(range(n_positions)):
        state, env = random_board_tris()
        evals = solver.evaluate_all_moves(env)

        max_eval = max(evals.values())
        if max_eval == 0:
            gto_result = 0
        else:
            gto_result = 1 if max_eval > 0 else -1
        # strong best moves: quickest win and slowest loss
        strong_best_moves = [move for move, val in evals.items() if val == max_eval]
        # weak best moves: quickness not counted, only result
        if gto_result == 0:
            weak_best_moves = strong_best_moves
        else:
            weak_best_moves = [move for move, val in evals.items() if val * gto_result > 0]
        sample = {
            "state": [state[0].tolist(), state[1]],
            "eval": evals,
            "strong_best_moves": strong_best_moves,
            "weak_best_moves": weak_best_moves,
            "gto_result": gto_result,
        }
        f.write(json.dumps(sample) + "\n")
        f.flush()
```

Connect4

```python
import bitbully as bb

from giotto.utils.bitbully_utils import convert_board
```

```python
solver = bb.BitBully()
```

```python
state, env = random_board_connect4()
env.render()
print(f"To move: {env.signs[env.current_player]}")

bb_board = convert_board(env.info["moves"])
print(bb_board)

scores = solver.score_all_moves(bb_board)

print("Move scores:", scores)
print(f"Best move: column {solver.best_move(bb_board) + 1}")
```

```python
n_positions = 10000
filename = "connect4_positions.jsonl"

with open(filename, mode="a") as f:
    for _ in tqdm(range(n_positions)):
        state, env = random_board_connect4()
        bb_board = convert_board(env.info["moves"])
        evals = solver.score_all_moves(bb_board)

        max_eval = max(evals.values())
        if max_eval == 0:
            gto_result = 0
        else:
            gto_result = 1 if max_eval > 0 else -1
        # strong best moves: quickest win and slowest loss
        strong_best_moves = [move + 1 for move, val in evals.items() if val == max_eval]
        # weak best moves: quickness not counted, only result
        if gto_result == 0:
            weak_best_moves = strong_best_moves
        else:
            weak_best_moves = [move + 1 for move, val in evals.items() if val * gto_result > 0]
        sample = {
            "state": [state[0].tolist(), state[1]],
            "eval": evals,
            "strong_best_moves": strong_best_moves,
            "weak_best_moves": weak_best_moves,
            "gto_result": gto_result,
        }
        f.write(json.dumps(sample) + "\n")
        f.flush()
```
