```python
from giotto.envs.connect4 import Connect4Env
from giotto.envs.tris import TrisEnv
```

Tris

```python
env = TrisEnv()
env.step(1)
# env.step(2)
env.render()

mock_targets = [1, 2, 3, 4, 5, 6, 7, 8, 9]
```

```python
augmented = env.simmetries.get_equivalent_boards(env.board, mock_targets)
print(len(augmented))
for aug in augmented:
    env.board = aug[0]
    env.render()
    print(aug[1])
```
